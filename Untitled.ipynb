```python
import pandas as pd
import os

# Directory containing CSV files
directory = r"C:\Users\Aniket Kumar\Downloads\painting museum data analysis"

# List of CSV files in the directory
csv_files = [file for file in os.listdir(directory) if file.endswith('.csv')]

# Initialize an empty list to store DataFrames
dfs = []

# Read each CSV file into a DataFrame and append it to the list
for file in csv_files:
    df = pd.read_csv(os.path.join(directory, file))
    dfs.append(df)

# Combine DataFrames into a single DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

# Write the combined DataFrame to a new CSV file
combined_df.to_csv('combined_csv_file.csv', index=False)
combined_df.head()
```

```python
import pandas as pd
import sqlite3

# Create a DataFrame (replace this with your DataFrame)
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35]}
df = pd.DataFrame(data)

# Connect to SQLite database (create it if it doesn't exist)
conn = sqlite3.connect('my_database.db')

# Convert the DataFrame to a SQL table
table_name = 'my_table'  # Specify the table name
df.to_sql(table_name, conn, if_exists='replace', index=False)

# Close the database connection
conn.close()

print(f'DataFrame has been converted to the SQL table: {table_name}')
```
